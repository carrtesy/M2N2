```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

## Univariate

Train

```python
plt.figure(figsize=(20, 6))
x = np.arange(10000)
p = (2*np.pi)/200
y = np.sin(p*x)
plt.plot(x, y)
```

```python
train = y[:, None]
train.shape
```

```python
train_label = np.zeros(10000)
train_label.shape
```

```python
with open('./toy_USW/train.npy', 'wb') as f:
    np.save(f, train)
with open('./toy_USW/train_label.npy', 'wb') as f:
    np.save(f, train_label)
```

```python
with open('./toy_USW/train.npy', 'rb') as f:
    train = np.load(f)
with open('./toy_USW/train_label.npy', 'rb') as f:
    train_label = np.load(f)
train.shape, train_label.shape
```

```python
plt.figure(figsize=(20, 6))
plt.plot(train, label = "data")
plt.plot(train_label, label = "anomaly label")
plt.legend()
plt.show()
print(train_label.sum())
```

Test

```python
plt.figure(figsize=(20, 6))
x = np.arange(9000)
p = (2*np.pi)/200
y = np.sin(p*x)
plt.plot(x, y)
```

```python
start = 0
y_test = []
anomaly_label = []
for i in range(10):
    a = np.random.choice(np.arange(start, start+1100))
    y_test += list(y[start:a])
    anomaly_label += [0] * len(y[start:a])
    
    
    print(len(y_test))
    y_test += ([y[a]] * 100)
    anomaly_label += [1] * (100)
    
    print(len(y_test))
    start = a
    print("+++")
y_test += list(y[start:])
anomaly_label += [0] * len(y[start:])

plt.figure(figsize=(20,6))
print(len(y_test), len(anomaly_label))
plt.plot(y_test)
plt.plot(anomaly_label)
plt.show()
```

```python
test = np.array(y_test)[:, None]
test.shape
```

```python
test_label = np.array(anomaly_label)
test_label.shape
```

```python
with open('./toy_USW/test.npy', 'wb') as f:
    np.save(f, test)
with open('./toy_USW/test_label.npy', 'wb') as f:
    np.save(f, test_label)
```

```python
with open('./toy_USW/test.npy', 'rb') as f:
    test = np.load(f)
with open('./toy_USW/test_label.npy', 'rb') as f:
    test_label = np.load(f)
test.shape, test_label.shape
```

```python
plt.figure(figsize=(20, 6))
plt.plot(test, label = "data")
plt.plot(test_label, label = "anomaly label")
plt.legend()
plt.show()
print(test_label.sum())
```
